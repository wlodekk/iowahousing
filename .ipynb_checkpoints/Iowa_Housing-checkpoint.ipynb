```python
def get_columns_from_transf(column_transformer):    
    col_name = []

    for transformer_in_columns in column_transformer.transformers_[:-1]: #the last transformer is ColumnTransformer's 'remainder'
        raw_col_name = transformer_in_columns[2]
        if isinstance(transformer_in_columns[1],Pipeline): 
            transformer = transformer_in_columns[1].steps[-1][1]
        else:
            transformer = transformer_in_columns[1]
        try:
            names = transformer.get_feature_names(raw_col_name)
        except AttributeError: # if no 'get_feature_names' function, use raw column name
            names = raw_col_name
        if isinstance(names,np.ndarray): # eg.
            col_name += names.tolist()
        elif isinstance(names,list):
            col_name += names    
        elif isinstance(names,str):
            col_name.append(names)

    [_, _, reminder_columns] = column_transformer.transformers_[-1]


    return col_name
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import TransformedTargetRegressor
from sklearn.linear_model import LinearRegression as lr
from sklearn.linear_model import Ridge
from sklearn.linear_model import HuberRegressor as hr
from sklearn.linear_model import Lasso
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline

prep_pipe = Pipeline([
    ('center',StandardScaler(with_std=False)),
    ('powertr',PowerTransformer(standardize=False)),
])

col_transf=ColumnTransformer([
    ('prep',prep_pipe,nums),
    ('scaler',RobustScaler(),nums)
])
pipe = Pipeline([
    ('transf',col_transf),
    ('est',lr())
])

params = [
    {
        'regressor__transf__prep__center':('passthrough',StandardScaler(with_std=False),),
        'regressor__transf__prep__powertr':('passthrough',PowerTransformer(standardize=False),),
        'regressor__transf__scaler':('passthrough',MinMaxScaler(),StandardScaler(),RobustScaler(),),
        'regressor__est':(lr(),),           
        'transformer':(None,MinMaxScaler(),StandardScaler(),RobustScaler(),)
    }, {
        'regressor__transf__prep__center':('passthrough',StandardScaler(with_std=False),),
        'regressor__transf__prep__powertr':('passthrough',PowerTransformer(standardize=False),),
        'regressor__transf__scaler':('passthrough',MinMaxScaler(),StandardScaler(),RobustScaler(),),
        'regressor__est':(Ridge(),),
        'transformer':(None,MinMaxScaler(),StandardScaler(),RobustScaler(),)
    }, {
        'regressor__transf__prep__center':('passthrough',StandardScaler(with_std=False),),
        'regressor__transf__prep__powertr':('passthrough',PowerTransformer(standardize=False),),
        'regressor__transf__scaler':('passthrough',MinMaxScaler(),StandardScaler(),RobustScaler(),),
        'regressor__est':(SVR(),),
        'transformer':(None,MinMaxScaler(),StandardScaler(),RobustScaler(),)
    },{
        'regressor__transf__prep__center':('passthrough',StandardScaler(with_std=False),),
        'regressor__transf__prep__powertr':('passthrough',),
        'regressor__transf__scaler':('passthrough',MinMaxScaler(),StandardScaler(),RobustScaler(),),
        'regressor__est':(hr(),), 
        'transformer':(None,MinMaxScaler(),StandardScaler(),RobustScaler(),)
    }, {
        'regressor__transf__prep__center':('passthrough',StandardScaler(with_std=False),),
        'regressor__transf__prep__powertr':('passthrough',PowerTransformer(standardize=False),),
        'regressor__transf__scaler':('passthrough',MinMaxScaler(),StandardScaler(),RobustScaler(),),
        'regressor__est':(Lasso(),),
        'transformer':(None,MinMaxScaler(),StandardScaler(),RobustScaler(),)
    }
]
ttr=TransformedTargetRegressor(regressor=pipe,transformer=RobustScaler())
grid_sr=GridSearchCV(ttr,params,scoring='neg_root_mean_squared_error',return_train_score=True,n_jobs=-1)
grid_sr.fit(enc_tr_data,log_tr_target)
```

```python
def results(grdsr):
    cv_res=grdsr.cv_results_
    df_res=pd.concat((pd.DataFrame(cv_res['mean_test_score'],columns=['neg_root_mean_squared_error']),pd.DataFrame(cv_res['std_test_score'],columns=['std_test_score']),pd.DataFrame(cv_res['params'])),axis=1)
    df_res=df_res.sort_values(by=['neg_root_mean_squared_error','std_test_score'],ascending=[False,True])
    return df_res
def top_res(grdsr,top_n=5):
    df_res=results(grdsr)
    df_res.iloc[:,2:]=df_res.iloc[:,2:].astype('string')
    df_res['max']=df_res.groupby('regressor__est')['neg_root_mean_squared_error'].transform(max)
    df_res_top_n=df_res.groupby('regressor__est').apply(lambda x: x.nlargest(top_n,'neg_root_mean_squared_error'))
    df_res_top_n=df_res_top_n.sort_values(['max','neg_root_mean_squared_error','std_test_score'],ascending=[False,False,True]).drop('max',axis=1)
    return df_res_top_n
    
```

```python
from sklearn.feature_selection import RFECV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import TransformedTargetRegressor
from sklearn.linear_model import LinearRegression as lr
from sklearn.linear_model import Ridge

from sklearn.pipeline import Pipeline

class PandasColumnTransformer(ColumnTransformer):
    def transform(self, X: pd.DataFrame) -> pd.DataFrame:
        return pd.DataFrame(super().transform(X), columns=X.columns, index=X.index)

    def fit_transform(self, X: pd.DataFrame, y=None) -> pd.DataFrame:
        return pd.DataFrame(super().fit_transform(X), columns=X.columns, index=X.index)


all_cat_ord=cat_ord+['LotShape']+['Utilities']+['LandSlope']+['ExterQual','ExterCond','HeatingQC','KitchenQual']+['BsmtQual','BsmtCond','FireplaceQu',
    'GarageQual','GarageCond']+['PoolQC']+['BsmtExposure']+['BsmtFinType1','BsmtFinType2']+['Functional']+['GarageFinish']

ridge_prep_pipe = Pipeline([
    ('center','passthrough'),
    ('powertr',PowerTransformer(standardize=False)),
])

ridge_col_transf= PandasColumnTransformer ([
    ('prep',ridge_prep_pipe,nums+all_cat_ord),
    ('scaler',MinMaxScaler(),nums+all_cat_ord)
])
ridge_pipe = Pipeline([
    ('transf',ridge_col_transf),
    ('est',Ridge(alpha=4))
])

ridge_ttr=TransformedTargetRegressor(regressor=ridge_pipe,transformer=MinMaxScaler())

ridge_rfecv=RFECV(ridge_ttr,min_features_to_select=180,cv=3,scoring='neg_root_mean_squared_error',n_jobs=-1,verbose=3)
ridge_rfecv.fit(enc_tr_data,log_tr_target)
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import TransformedTargetRegressor
from sklearn.linear_model import LinearRegression as lr
from sklearn.linear_model import Ridge
from catboost import CatBoostRegressor 
from sklearn.pipeline import Pipeline

all_cat_ord=cat_ord+['LotShape']+['Utilities']+['LandSlope']+['ExterQual','ExterCond','HeatingQC','KitchenQual']+['BsmtQual','BsmtCond','FireplaceQu',
    'GarageQual','GarageCond']+['PoolQC']+['BsmtExposure']+['BsmtFinType1','BsmtFinType2']+['Functional']+['GarageFinish']

ridge_prep_pipe = Pipeline([
    ('center','passthrough'),
    ('powertr',PowerTransformer(standardize=False)),
])

ridge_col_transf=ColumnTransformer([
    ('prep',ridge_prep_pipe,nums+all_cat_ord),
    ('scaler',MinMaxScaler(),nums+all_cat_ord)
])
ridge_pipe = Pipeline([
    ('transf',ridge_col_transf),
    ('est',Ridge(alpha=4))
])

lr_prep_pipe = Pipeline([
    ('center',StandardScaler(with_std=False)),
    ('powertr',PowerTransformer(standardize=False)),
])

lr_col_transf=ColumnTransformer([
    ('prep',lr_prep_pipe,nums+all_cat_ord),
    ('scaler',RobustScaler(),nums+all_cat_ord)
])
lr_pipe = Pipeline([
    ('transf',lr_col_transf),
    ('est',lr())
])

ridge_ttr=TransformedTargetRegressor(regressor=ridge_pipe,transformer=MinMaxScaler())
lr_ttr=TransformedTargetRegressor(regressor=lr_pipe,transformer=RobustScaler())
cbr_model= CatBoostRegressor(task_type='GPU',devices='0',verbose=False,n_estimators=400,depth=4,l2_leaf_reg=1,learning_rate=0.06,subsample=0.9,random_seed=0)

stack_estimators=[
    ('ridge',ridge_ttr),
    ('lr','drop'),
    ('cbr',cbr_model)
                 ]

stack_reg=StackingRegressor(estimators=stack_estimators,final_estimator=lr())
params = [
     {
        'final_estimator':(lr(),Ridge(),),
        'passthrough':(False,True,),
        #'estimators__lr':(lr_ttr,'drop')
    }
]

grid_sr=GridSearchCV(stack_reg,params,scoring='neg_root_mean_squared_error',return_train_score=True,n_jobs=1)

grid_sr.fit(enc_tr_data,log_tr_target)

```
