```python
from sklearn.pipeline import Pipeline
from sklearn.compose import  ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
#from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
from sklearn.model_selection import GridSearchCV

num_pipe= Pipeline([
    ('imputer',SimpleImputer()),
    ('scaler',StandardScaler())
])
cat_pipe=Pipeline([
    ('imputer',SimpleImputer()),
    ('encoder',OneHotEncoder())
])

nums=['LotFrontage','LotArea','YearBuilt','YearRemodAdd','MasVnrArea','smtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF',
      '1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','Bedroom','Kitchen','TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea',
      'WoodDeckSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold']

cat_ord=['OverallQual','OverallCond','CentralAir']
cat=['MSSubClass','MSZoning','Street','Alley','LandContour','Neighborhood','Condition1','Condition2','BldgType',
     'HouseStyle','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','Foundation','Heating','GarageType','GarageFinish','PavedDrive','Fence','MiscFeature','SaleType','SaleCondition']
num_pipe=Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value=0)),('scaler',RobustScaler())])
prep_pipe=ColumnTransformer([
    ('num',num_pipe,nums),
    ('cat_ord',OrdinalEncoder(),cat_ord),
    ('cat_hot',Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value='None')),('encoder',OneHotEncoder())]),cat),
    ('Lotsh',OrdinalEncoder(categories=[['IR3','IR2','IR1','Reg']]),'LotShape'),
    ('Util',OrdinalEncoder(categories=[['Elo','NoSeWa','NoSewr','AllPub']]),'Utilities'),
    ('Ext',OrdinalEncoder(categories=[['Po','Fa','TA','Gd','Ex']]),['ExterQual','ExterCond','HeatingQC','KitchenQual']),
    ('QualCond',Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value='None')),('encoder',OrdinalEncoder(categories=[['None','Po','Fa','TA','Gd','Ex']]))]),['BsmtQual','BsmtCond','FireplaceQu',
    'GarageQual','GarageCond']),
    ('PoolCond',Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value='None')),('encoder',OrdinalEncoder(categories=[['None','Fa','TA','Gd','Ex']]))]),'PoolCond'),
    ('BsmtExp',Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value='None')),('encoder',OrdinalEncoder(categories=[['None','No','Mn','Av','Gd']]))]),['BsmtExposure']),
    ('BsmtFinType',Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value='None')),('encoder',OrdinalEncoder(categories=[['None','Unf','LwQ','BLQ','ALQ','GLQ']]))]),['BsmtFinType1','BsmtFinType2']),
    ('Functional',OrdinalEncoder(categories=[['Sal','Sev','Maj2',' Maj1','Mod','Min2','Min1','Typ']]),'Functional'),
    ('MasVnrType',Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value='None')),('encoder',OneHotEncoder())]),'MasVnrType'),
    ('GarageFinish',Pipeline([('imputer',SimpleImputer(strategy='constant',fill_value='None')),('encoder',OrdinalEncoder(categories=[['None','Unf','RFn','Fin']]))]),['BGarageFinish'])
    ('Electrical',Pipeline([('imputer',SimpleImputer(strategy='most_frequent')),('encoder',OneHotEncoder())]),'Electrical')
])
     
                   
                  
```
